```python
dataset_prep = dataset.copy() 
#Now I have taken a copy of my census income dataset to maintain the cleaned one that is used to later,
## and also to use the copied one for preparing the data for the model.
```

Now i am going to scaling which means:-
This is a technique to standardize the independent variables present in the data in a fixed range. It is performed during the data pre-processing to handle highly varying values or units. So, I use Feature Scaling to bring all values to the same magnitudes, like to say,to tackle the issues 

```python
# Scalling
from sklearn.preprocessing import MinMaxScaler
numerical = ['age', 'capital_gains', 'capital_losses', 'wages', 'weeks_worked']

scaler = MinMaxScaler()
dataset_prep[numerical] = scaler.fit_transform(dataset_prep[numerical])
```

```python
dataset_prep.sample(3)##returns a list with a randomly selection of a specified number of items from a sequnce
```

Now, my data has been scaled to MinMaxScalling for numerical attributes, which converts the data to have a range between 0 and 1. That would help to make the data well-prepared for the model.


```python
# Encoding : This is the first method to convert categorical values in numeric.
dataset_prep['sex'] = dataset_prep.sex.replace({"Female": 0, "Male": 1})
dataset_prep['income_50k'] = dataset_prep.income_50k.replace({"- 50000": 0, "50000+": 1})

```

```python
# Create dummy variables: this is the easiest way to convert categorical to numeric data.
dataset_prep = pd.get_dummies(dataset_prep) #converts categorical data into dummy or indicator variables
```

```python
# import some classification models
from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

# import needed functions

from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

```python
#Separating the independent variables and dependent variables
x = dataset_prep.iloc[:,:-1]
y = dataset_prep.iloc[:,-1]
```

```python
# random state: 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=3)
#The data has been split into training and testing parts of the features and 
#the label with a test size of 30% and without a random state with the next runs.
# This happened by using the train_test_split function.
```

```python
x_train.head(3)
```

```python
#working with SMOTE library : This algorithm helps to overcome the overfitting problem posed by oversampling
from imblearn.over_sampling import SMOTE    #SMOTE - Synthetic Minority Over-sampling Technique
SMOTE = SMOTE()
```

```python
#fit and apply the transform
x_train_SMOTE,y_train_SMOTE = SMOTE.fit_resample(x_train, y_train)

```

```python
print("after SMOTE")
y_train_SMOTE.value_counts()
```

```python
x_train_SMOTE.duplicated().sum()
```

```python
#Models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

Logistic regression, is a linear model for classification. In this model, the probabilities describing the possible outcomes of a single trial are modeled using a logistic function.

```python
model=LogisticRegression()
model.fit(x_train_SMOTE,y_train_SMOTE)#takes the training data as arguments
predict_logistic=model.predict(x_test)
predict_logistic

```

```python
y_test
```

```python
confusion_matrix_logistic=confusion_matrix(y_test,predict_logistic)#takes in the list of actual labels, the list of predicted labels, and an optional argument to specify the order of the labels
logistic_acc=accuracy_score(y_test,predict_logistic)*100
print("accuracy of logistic regression:",round(logistic_acc,2),"%")
```

Confusion metrics is usedto evaluate the performance of a classification model through the calculation of performance metrics like accuracy

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
model = knn.fit(x_train_SMOTE,y_train_SMOTE)
pred_k=model.predict(x_test)
knn_acc=accuracy_score(y_test,pred_k)*100
print("acuracy of KNN:", round(knn_acc,2),"%")
```

```python
from sklearn.naive_bayes import GaussianNB
GNB=GaussianNB()
model2=GNB.fit(x_train_SMOTE,y_train_SMOTE)
pred_GNB=model2.predict(x_test)
GNB_acc=accuracy_score(y_test,pred_GNB)*100
print("accuracy of KNN:",round(GNB_acc,2),"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()
model3=RFC.fit(x_train_SMOTE,y_train_SMOTE)
pred_RFC=model3.predict(x_test)
RFC_acc=accuracy_score(y_test,pred_RFC)*100
print("accuracy of knn:",round(RFC_acc,2),"%")
```

As a result, I observe that Random Forest Classifier over fits the train set, and gives a 99% accuracy on train data.
